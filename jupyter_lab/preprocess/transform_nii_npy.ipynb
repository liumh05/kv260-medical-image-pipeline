```python
import nibabel as nib
import numpy as np

# 读取文件
nii_path = 'imaging.nii.gz'
img = nib.load(nii_path)

# 提取数据和仿射矩阵
data = img.get_fdata()
affine = img.affine

# 保存为.npy文件
np.save('data.npy', data)          # 图像数据
np.save('affine.npy', affine)      # 仿射矩阵

print("数据形状:", data.shape)
print("仿射矩阵:\n", affine)
```
